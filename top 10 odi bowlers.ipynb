```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://www.icc-cricket.com/rankings/mens/player-rankings/odi/bowling'
```

```python
page = requests.get(url)
page
```

```python
page.content
```

```python
soup = BeautifulSoup(page.content, "html.parser")
print(soup.prettify())
```

```python
bowler_rank = soup.find_all('tr',class_="table-body")
```

```python
len(bowler_rank)
```

```python
bowler_name = []
team_name = []
rating = []

for i in bowler_rank:
    
    bowler_name.append(i.find('td',class_='table-body__cell rankings-table__name name').text.strip())
    team_name.append(i.find('td',class_='table-body__cell nationality-logo rankings-table__team').text.strip())
    rating.append(i.find('td' ,class_='table-body__cell rating').text.strip())
```

```python
bowler_name
```

```python
team_name
```

```python
rating
```

```python
d={'bowler_name':bowler_name, 'team_name':team_name, 'rating':rating}

df = pd.DataFrame(d)

df
```
