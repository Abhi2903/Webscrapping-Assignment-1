```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/Main_Page'
```

```python
page = requests.get(url)
page
```

```python
page.content
```

```python
soup = BeautifulSoup(page.content, "html.parser")
print(soup.prettify())
```

```python
header_tag = soup.find_all('h2',class_="mp-h2")
```

```python
header_tag
```

```python
headline = []

for i in header_tag:
    
      headline.append(i.find('span',class_='mw-headline').text)
```

```python
headline
```

```python
d={'headline':headline}

df = pd.DataFrame(d)

df.head(10)
```
